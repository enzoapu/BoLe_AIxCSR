```python
import pandas as pd
import numpy as np
train = pd.read_csv('./train.csv')
```

```python
df = train.sample(n=100)

df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Creating the dependent variable class
factor = pd.factorize(df['Activity'])
df.Activity = factor[0]
definitions = factor[1]
print(df.Activity.head())
print(definitions)
```

```python

#Splitting the data into independent and dependent variables
X = df.iloc[:,0:-1].values
y = df.iloc[:,-1].values
print('The independent features set: ')
print(X[:5,:])
print('The dependent variable: ')
print(y[:5])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 21)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
#Reverse factorize (converting y_pred from 0s,1s and 2s to Iris-setosa, Iris-versicolor and Iris-virginica
reversefactor = dict(zip(range(6),definitions))
y_test = np.vectorize(reversefactor.get)(y_test)
y_pred = np.vectorize(reversefactor.get)(y_pred)
# Making the Confusion Matrix
print(pd.crosstab(y_test, y_pred, rownames=['Actual Species'], colnames=['Predicted Species']))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(classifier, X, y, cv=5)
print(scores)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
def print_accuracy_report(classifier, X, y, num_validations=5):
    accuracy = cross_val_score(classifier, 
            X, y, scoring='accuracy', cv=num_validations)
    print("Accuracy: " + str(round(100*accuracy.mean(), 2)) + "%")

    f1 = cross_val_score(classifier, 
            X, y, scoring='f1_weighted', cv=num_validations)
    print("F1: " + str(round(100*f1.mean(), 2)) + "%")

    precision = cross_val_score(classifier, 
            X, y, scoring='precision_weighted', cv=num_validations)
    print("Precision: " + str(round(100*precision.mean(), 2)) + "%")

    recall = cross_val_score(classifier, 
            X, y, scoring='recall_weighted', cv=num_validations)
    print("Recall: " + str(round(100*recall.mean(), 2)) + "%" )

def trainResult(df, classifier):
    factor = pd.factorize(df['Activity'])
    df.Activity = factor[0]
    definitions = factor[1]

    X = df.iloc[:,0:-1].values
    y = df.iloc[:,-1].values

#     print(X, y)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 21)

    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

#     classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
    classifier.fit(X_train, y_train)

    y_pred = classifier.predict(X_test)
    reversefactor = dict(zip(range(6),definitions))
    y_test = np.vectorize(reversefactor.get)(y_test)
    y_pred = np.vectorize(reversefactor.get)(y_pred)

    # Making the Confusion Matrix
#     print(pd.crosstab(y_test, y_pred, rownames=['Actual Species'], colnames=['Predicted Species']))

    scores = print_accuracy_report(classifier, X, y)
    print()
    return
```

```python
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
print('n = 100, RandomForestClassifier')
trainResult(train.sample(n=100), classifier)
print('n = 500, RandomForestClassifier')
trainResult(train.sample(n=500), classifier)
print('n = 1000, RandomForestClassifier')
trainResult(train.sample(n=1000), classifier)
print('n = 3000, RandomForestClassifier')
trainResult(train.sample(n=3000), classifier)
print('n = 5000, RandomForestClassifier')
trainResult(train.sample(n=5000), classifier)
print('n = all, RandomForestClassifier')
trainResult(train, classifier)
```

```python
classifier = OneVsRestClassifier(KNeighborsClassifier(),-1)
print('n = 100, KNeighborsClassifier')
trainResult(train.sample(n=100), classifier)
print('n = 500, KNeighborsClassifier')
trainResult(train.sample(n=500), classifier)
print('n = 1000, KNeighborsClassifier')
trainResult(train.sample(n=1000), classifier)
print('n = 3000, KNeighborsClassifier')
trainResult(train.sample(n=3000), classifier)
print('n = 5000, KNeighborsClassifier')
trainResult(train.sample(n=5000), classifier)
print('n = all, KNeighborsClassifier')
trainResult(train, classifier)
```

```python
classifier = OneVsRestClassifier(LogisticRegression(),-1)
print('n = 100, LogisticRegression')
trainResult(train.sample(n=100), classifier)
print('n = 500, LogisticRegression')
trainResult(train.sample(n=500), classifier)
print('n = 1000, LogisticRegression')
trainResult(train.sample(n=1000), classifier)
print('n = 3000, LogisticRegression')
trainResult(train.sample(n=3000), classifier)
print('n = 5000, LogisticRegression')
trainResult(train.sample(n=5000), classifier)
print('n = all, LogisticRegression')
trainResult(train, classifier)
```

```python
classifier = OneVsRestClassifier(MLPClassifier(),-1)
print('n = 100, MLPClassifier')
trainResult(train.sample(n=100), classifier)
print('n = 500, MLPClassifier')
trainResult(train.sample(n=500), classifier)
print('n = 1000, MLPClassifier')
trainResult(train.sample(n=1000), classifier)
print('n = 3000, MLPClassifier')
trainResult(train.sample(n=3000), classifier)
print('n = 5000, MLPClassifier')
trainResult(train.sample(n=5000), classifier)
print('n = all, MLPClassifier')
trainResult(train, classifier)
```
